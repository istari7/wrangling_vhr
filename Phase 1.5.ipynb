### Phase 1.5 
#### Combining raw and cleaned subsets to get a proper partition

```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_excel("raw.xlsx")
clean_data = pd.read_excel("new_clean.xlsx")
```

```python
df1 = pd.DataFrame(raw_data)
df2 = pd.DataFrame(clean_data)
raw_df = df1.iloc[:10000,:]
```


## Initial Cleaned Data (Email, Name, Location)
## We get this from cleaning the previous 20k entry file for email. That only had location cleaned. Not emails
## Length ---> 15595

```python
df2.shape
```

## Completely clean and usable dataset after merging
## Length ---> 15587

```python
perfect_clean_df = pd.merge(df1, df2, how ='inner', indicator=False, on =["Email"])
```

```python
perfect_clean_df.shape
```

## Data that has to be cleaned for the location. Heuristics or Paid? Pending Confirmation
## Length ----> 79175

```python
unclean_data = df1.merge(df2, how = 'outer' ,indicator=True,on=["Email"]).loc[lambda x : x['_merge']=='left_only']
```

```python
unclean_data.shape
```

### Cleaned Location Data not in raw dataset, also lacks additional metadata

```python
clean_but_not_match = df1.merge(df2, how = 'outer' ,indicator=True,on=["Email"]).loc[lambda x : x['_merge']=='right_only']
```

```python
perfect_clean_df.to_excel("perfect.xlsx")
unclean_data.to_excel("semi_raw.xlsx")
```
